```python
sex = df['Sex']
survived = df['Survived']

count = 0
total = 0

for s, sv in zip(sex, survived):
    if pd.isna(s) or pd.isna(sv):
        continue

    total += 1

    if str(s) != str(sv):
        count += 1

dissimilarity = count / total

print("Nominal Attribute Dissimilarity")
print("Sex vs Survived =", round(dissimilarity, 4))
```
